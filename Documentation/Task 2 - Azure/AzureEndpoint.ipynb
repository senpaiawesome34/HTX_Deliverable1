```python
# import the modules required
import os

# run in a notebook cell before creating AzureOpenAI
os.environ["AZURE_OPENAI_API_KEY"] = "YOUR_KEY_HERE"
os.environ["OPENAI_API_VERSION"] = "YOUR_API_VERSION_HERE"   # e.g. from Azure docs/portal
os.environ["AZURE_OPENAI_ENDPOINT"] = "https://YOUR_RESOURCE_NAME.openai.azure.com/"
os.environ["AZURE_OPENAI_DEPLOYMENT_NAME"] = "YOUR_DEPLOYMENT_NAME"
```

Ensure:

1. Endpoint must end with .openai.azure.com/
2. Deployment name must be the one you created in Azure AI Foundry/Azure OpenAI deployments
3. API version must match a valid Azure OpenAI version
4. If you set env vars in PowerShell, restart notebook kernel or set again in notebook

```python
from getpass import getpass
import os

os.environ["AZURE_OPENAI_API_KEY"] = getpass("AZURE_OPENAI_API_KEY: ")
os.environ["OPENAI_API_VERSION"] = input("OPENAI_API_VERSION: ").strip()
os.environ["AZURE_OPENAI_ENDPOINT"] = input("AZURE_OPENAI_ENDPOINT: ").strip()
os.environ["AZURE_OPENAI_DEPLOYMENT_NAME"] = input("AZURE_OPENAI_DEPLOYMENT_NAME: ").strip()
```

```python
required_env = [
    'AZURE_OPENAI_API_KEY',
    'OPENAI_API_VERSION',
    'AZURE_OPENAI_ENDPOINT',
    'AZURE_OPENAI_DEPLOYMENT_NAME',
]

missing = [name for name in required_env if not os.environ.get(name)]
if missing:
    raise EnvironmentError(f"Missing environment variable(s): {', '.join(missing)}")

client = AzureOpenAI(
    api_key=os.environ['AZURE_OPENAI_API_KEY'],
    api_version=os.environ['OPENAI_API_VERSION'],
    azure_endpoint=os.environ['AZURE_OPENAI_ENDPOINT']
)

resp = client.chat.completions.create(
    model=os.environ['AZURE_OPENAI_DEPLOYMENT_NAME'],  # deployment name, not raw model id
    messages=[
        {'role': 'system', 'content': 'You are a helpful assistant'},
        {'role': 'user', 'content': 'Summarize Azure resource groups in 3 bullets'}
    ]
)

print(resp.choices[0].message.content)
```
